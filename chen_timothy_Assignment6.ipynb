```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns 
    
from sklearn import metrics
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
# configuring dataframe
bos = pd.DataFrame(boston.data)
bos.columns = boston.feature_names
# adding target (price) to the dataframe
bos['PRICE'] = boston.target

bos.head()
```

```python
# declare inputs and target variables
target = bos['PRICE']
inputs = bos.drop(['PRICE'], axis=1)
```

```python
errors = []
scores = []

for i in range(10):
    # train test split

    X_train, X_test, Y_train, Y_test = train_test_split(inputs, target, shuffle=True, test_size=0.2)


    # create regression model


    # from sklearn.metrics import r2_score, mean_squared_error

    # Create a linear regression object
    reg = LinearRegression()

    # Fit the regression with the scaled TRAIN inputs and targets
    reg.fit(X_train, Y_train)

    # store outputs of regression in Y_hat
    Y_hat = reg.predict(X_train)

    # compare the target (Y_train) and the prediction via plot
   # plt.plot(Y_train, Y_hat, 'o', color='green')

    # regression line
    m, b = np.polyfit(Y_train, Y_hat, 1)

    # axis labels
    #plt.xlabel('Targets (Y_train)',size=18)
    #plt.ylabel('Predictions (Y_hat)',size=18)

#    colors = {'Y_train': 'red', 'Y_hat': 'green'}

    # add linear regression line to plot
#    plt.plot(Y_train, m*Y_train+b)
    
 #   plt.show()
# Several statistics can be generated from a linear model. Given a fitted linear model, the following code outputs the model 
# coefficients (the parameter values for the fitted model), the residual sum of squares (the model error), and the 
# explained variance or R square or Variance Score(the degree to which the model explains the variation present in the data)

    print("Iteration : ", i + 1)
    # The coefficients
    print('Coefficients:', reg.coef_)
    # The mean squared error
    mse = np.mean((reg.predict(X_test) - Y_test) ** 2)
    print("Mean squared error: %.2f" % mse )
    # Explained variance score: 1 is perfect prediction
    score = reg.score(X_test, Y_test)
    print('Variance score: %.2f' % score)
    scores.append(score)
    errors.append(mse)
    print()
    
print("Overall");
print("MSE {:0.2f}".format(sum(errors) / 10))
print("Variance {:0.2f}".format(sum(scores) / 10))
```

```python
# keep track of the MSE and the regression score for each column using this dictionary
# both the values for each volumn with initially be 0
# during each iteration keep a running total by adding the mse and the regression score
# to the two values held in the dictionary (the dictionary value is a list of 2)
d = {x: [0 , 0]  for x in X_train.columns}

for j in range(10):
    X_train, X_test, Y_train, Y_test = train_test_split(inputs, target, shuffle=True, test_size=0.2) #, random_state=0)
    for i,col in enumerate(X_train.columns):
        print("**************************************")
        print("ITERATION #", j + 1)
        print(f"analysing for column -> {col}")
        # Create a linear regression object
        reg = LinearRegression()

        fig = plt.gcf();
        fig.set_size_inches(12, 8)
        plt.subplot(5, 3, i+1)
        # Fit the regression with the scaled TRAIN inputs and targets
        reg.fit(X_train[[col]], Y_train)
        Y_hat = reg.predict(X_train[[col]])
        Y_pred = reg.predict(X_test[[col]])

        colors = {'Y_train': 'red', 'Y_hat': 'green'}
        # compare the target (Y_train) and the prediction via plot
        plt.plot(X_train[col], Y_hat, 'o', color='green') # predict y for the train set
        plt.plot(X_train[col], Y_train, '^', color='red') # actual y for the train set
        plt.plot(X_test[col], Y_pred, 'v', color='green') # predicted y for the test set
        plt.plot(X_test[col], Y_test, '*', color='red') # actual y for the test set

        # regression line
        m, b = np.polyfit(X_train[col], Y_hat, 1)

        # axis labels
        plt.xlabel('Targets (Y_train)',size=10)
        plt.ylabel('Predictions (Y_hat)',size=10)

        # add linear regression line to plot
        plt.plot(X_train[col], m * X_train[col]+b)


        # The coefficients
        print('Coefficients:', reg.coef_)
        # The mean squared error
        mse = np.mean( (Y_pred - Y_test) ** 2)
        print("Mean squared error: %.2f" % mse )
        # Explained variance score: 1 is perfect prediction
        reg = reg.score(X_test[[col]], Y_test)
        print('Variance score: %.2f' % reg)
    
        d[col][0] += mse;
        d[col][1] += reg

        plt.show();
```

```python
# add the score and mse for the overall into our dictionary
for key, val in d.items():
    val[0] = val[0]/10
    val[1] = val[1] / 10
    
d['all'] = [sum(errors) / 10, sum(scores) / 10]
d['all']
```
